```python
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np

from sklearn.datasets import make_swiss_roll, make_blobs
from sklearn import decomposition
from sklearn.manifold import TSNE
```

```python
def generate_swiss_roll_data(n_samples):
    noise = 0.05
    X, _ = make_swiss_roll(n_samples, noise)
    # Make it thinner
    X[:, 1] *= .5
    distance_from_y_axis = X[:, 0] ** 2 + X[:, 2] ** 2 
    X_color = plt.cm.jet(distance_from_y_axis / np.max(distance_from_y_axis + 1))
    return X, X_color


centers = [(10, 9, 4), (4, 2, 9), 
           (9, 3, 9), (10, 8, 10), 
           (6,2,4), (3, 9, 11), 
           (1, 3, 7)]

def generate_gaussian_blobs_different(n_samples):
    X, Y = make_blobs(n_samples=n_samples,centers = centers, n_features=3)
    X[Y == 2, :] *= 1.5
    
    X_color = plt.cm.jet_r(Y / np.max(Y + 1), 0.1)
    Y = list(Y.astype(str))
    Y = list(map(lambda x: 'Cell_type_'+x, Y))
    
    return X, Y

X, Y = generate_gaussian_blobs_different(2500)
```

```python

def generate_gaussian_blobs_different_flat(n_samples):
    X, Y = make_blobs(n_samples=n_samples,centers = [(0, 9, 0), (0, 2, 0), 
                                                     (0, 3, 0), (0, 8, 0), 
                                                     (0, 2, 0), (0, 9, 0), 
                                                     (0, 3, 0)]
                      , n_features=3)
    X[Y == 1, :] *= 1.5
    
    X_color = plt.cm.jet_r(Y / np.max(Y + 1), 0.1)
    Y = list(Y.astype(str))
    Y = list(map(lambda x: 'Cell_type_'+x, Y))
    
    X[:,0] *= 0
    X[:,2] *= 0
    return X, Y

X, Y = generate_gaussian_blobs_different_flat(2500)
```

```python
fig = px.scatter_3d(x = X[:,0], y = X[:,1], z = X[:,2],
                   color = Y,
                   color_discrete_map = {'Cell_type_0': 'rgb(255, 100, 255)',
                                        'Cell_type_1': 'rgb(58, 220, 100)',
                                        'Cell_type_2': 'rgb(100, 255, 255)'}
                 )
fig.update_traces(marker=dict(size=1))
fig.update_traces(hovertemplate=None, hoverinfo='skip')

fig.update_layout(scene = dict(
                    xaxis = dict(
                         visible= False,
                         backgroundcolor="rgb(255, 255, 255)"),
                    yaxis = dict(
                         visible= False,
                         backgroundcolor="rgb(255, 255, 255)"),
                    zaxis = dict(
                         visible= False,
                         backgroundcolor="rgb(255, 255, 255)")
                            ) 
                  )
fig.update_layout(width=1000, margin=dict(r=10, l=10, b=10, t=10))

#fig.update_layout(scene_aspectmode='cube')

axis1 = go.Scatter3d(x=[-1,15], y=[0,0], z=[0, 0], mode = 'lines',
                     line = dict(color = 'rgb(0, 0, 0)', width= 1), showlegend=False)
axis2 = go.Scatter3d(x=[0,0], y=[-1,15], z=[0, 0], mode = 'lines',
                     line = dict(color = 'rgb(0, 0, 0)', width= 1), showlegend=False)
axis3 = go.Scatter3d(x=[0,0], y=[0,0], z=[-1,15], mode = 'lines',
                     line = dict(color = 'rgb(0, 0, 0)', width= 1), showlegend=False)

fig.add_trace(axis1)
fig.add_trace(axis2)
fig.add_trace(axis3)
```

```python
X_pca = decomposition.PCA(n_components=2).fit_transform(X)

```

```python
fig = px.scatter(x = X_pca[:,0], y = X_pca[:,1],
                   color = Y,
                   color_discrete_map = {'0': 'rgb(255, 100, 255)',
                                        '1': 'rgb(255, 255, 100)',
                                        '2': 'rgb(100, 255, 255)'}
                   )
fig.show()
```

```python
X_tsne = TSNE(n_components=2).fit_transform(X)
```

```python
fig = px.scatter(x = X_tsne[:,0], y = X_tsne[:,1],
                   color = Y,
                   color_discrete_map = {'0': 'rgb(255, 100, 255)',
                                        '1': 'rgb(255, 255, 100)',
                                        '2': 'rgb(100, 255, 255)'}
                   )
fig.show()

## Map down into 2d or 1d 
```
